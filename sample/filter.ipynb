```python
import numpy as np
import cv2
from matplotlib import pyplot as plt
```

```python
img=cv2.imread('girl.jpg')

plt.imshow(img)
```

```python
kernel=np.ones((5,5),np.float32)/25
dst= cv2.filter2D(img,-1,kernel)
plt.imshow(dst); plt.show()
```

```python
dst=cv2.GaussianBlur(img,(3,3),-1)
plt.imshow(dst)
```
